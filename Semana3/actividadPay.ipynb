```python
#Crear y guardar una gráfica de pay del año más reciente existente en la base de datos.
import numpy as np
import pandas as pd
data_atletas = pd.read_csv("C:\\Users\\mespinosa\\Documents\\pythonFundamental\\Semana3\\athlete_events.csv")
data_atletas
```

```python
dataMedallas = data_atletas[["NOC","Year","Medal"]]
dataClean = dataMedallas.dropna()
lastYear = data_atletas["Year"].max()
lastYear
```

```python
data_medallas2=dataClean[["NOC","Year"]]
data_medallas2
```

```python
dataLastYear2 = data_medallas2[data_medallas2["Year"] == lastYear]
dataLastYear2
```

```python
dataLastYear2["NOC"].value_counts()
```

```python
nocList=  dataLastYear2["NOC"].unique()
nocList.tolist()
```

```python
dataNoc = dataLastYear2["NOC"].value_counts()
```

```python
pieChar = dataNoc.plot.pie(figsize=(10,10),labels = nocList,autopct="%0.5f %%")
```

```python
fig = pieChar.get_figure()
```

```python
fig.savefig("Grafica_pay_2016_Paises.png")
```

```python
dataLastYear = dataClean[dataClean["Year"] == lastYear]
```

```python
dataLastYear
```

```python
dataDummy = pd.get_dummies(dataLastYear["Medal"])
```

```python
dataDummy
```

```python
dataConcatenada = pd.concat([dataLastYear,dataDummy],axis=1)
dataConcatenada
```

```python
dataAgrupada = dataConcatenada.groupby(["Year","Medal"]).sum()
dataAgrupada
```

```python
dataAgrupada["TotalMedallas"] = dataAgrupada["Bronze"] + dataAgrupada["Gold"] + dataAgrupada["Silver"]
```

```python
dataAgrupada
```

```python
pieChar2 = dataAgrupada["TotalMedallas"].plot.pie(figsize=(11,6),autopct="%0.5f %%")
```

```python
fig2 = pieChar2.get_figure()
fig2.savefig("Grafica_pay_2016_Medallas.png")
```
